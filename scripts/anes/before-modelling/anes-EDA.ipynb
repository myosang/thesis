# Exploratory Data Analysis

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# mentions
mentions_stack= pd.read_json("~/thesis/data/processed_uscensus/political_mentions_stack.jsonl", orient = 'records', lines = True)
mention_most = pd.read_json("~/thesis/data/processed_uscensus/political_mention_most.jsonl", orient = 'index')
```

```python
# dataframe of mentions that 1,2,3 mentions are all stacked
print(mentions_stack['stack'].head(3))
```

```python
# dataframe of mention that is the most important from the participant's 1,2,3 mentions
print(mention_most['mention_most'].head(3))
```

## WordCloud

### 1. mentions 1,2,3 all stack

```python
# Import the wordcloud library
from wordcloud import WordCloud

# Join the different processed titles together.
long_string = ','.join(list(mentions_stack['stack'].values))

# Create a WordCloud object
wordcloud = WordCloud(width = 500, height = 400, background_color="white", max_words=5000, contour_width=3, contour_color='steelblue')

# Generate a word cloud
wordcloud.generate(long_string)

# Visualize the word cloud
wordcloud.to_image()
```

### 2. most important mention by participant

```python
# Join the different processed titles together.
long_string = ','.join(list(mention_most['mention_most'].values))

# Create a WordCloud object
wordcloud = WordCloud(width = 500, height = 400, background_color="white", max_words=5000, contour_width=3, contour_color='steelblue')

# Generate a word cloud
wordcloud.generate(long_string)

# Visualize the word cloud
wordcloud.to_image()
```
