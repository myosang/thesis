### Open-ended responses from ANES 2020

```python
import pandas as pd

# file path
file_path = '~/thesis/data/open-anes-2020/anes_timeseries_2020_redactedopenends_excel_20211118.xlsx'

# Load all sheets of the Excel file into a dictionary
sheets_dict = pd.read_excel(file_path, sheet_name=None)

# Now sheets_dict contains all the sheets, where keys are sheet names and values are DataFrames
# You can access each DataFrame by its sheet name
for sheet_name, df in sheets_dict.items():
    print(f"Sheet: {sheet_name}")
    print(df.head())  # This will print the first few rows of each DataFrame
```

```python
# select the columns for analysis
# open-ended responses sheets names = ['V202205', 'V202207', 'V202209', 'V202211']
first = sheets_dict['V202205']
second = sheets_dict['V202207']
third = sheets_dict['V202209']
most = sheets_dict['V202211']

from functools import reduce

# compile the list of dataframes you want to merge
data_frames = [first, second, third, most]
df = reduce(lambda  left, right: pd.merge(left,right,on=['V200001'], # reduce function from the functools
                                            how='outer'), data_frames)
df
```

```python
df.to_csv('~/thesis/data/processed_data/anes_open_responses.csv', index = False)
```

```python
# compile the list of dataframes to merge
data_frames = [df1, df2, df3, dfmost]
df = reduce(lambda  left, right: pd.merge(left,right,on=['id_case'], # reduce function from the functools
                                            how='outer'), data_frames)
```

### cleaning each column dataframe
1. non-english words removal
2. stopwords removal
3. lemmatization
4. null value('empty string') removal

```python
# text cleaning & stopwords removal
import nltk
import re
from nltk.corpus import stopwords

# Lemmatizer modules
from nltk.stem import WordNetLemmatizer
```

```python
df
```

### ['mention1']

```python
# define the data from each dataframe columns for preprocessing
data = df1['mention1'].copy()
# define the id case that are not null for preprocessing
ind_list = df1['id_case'].dropna(axis = 0, how = 'any').to_list()
```

```python
# Remove special characters, lower the cases, remove stopwords.

# dictionary to contain
new_dict = {case: [] for case in ind_list}

# preprocess - remove special character, lower case, drop stopwords
for i in range(len(data)): 
    
    text = data.iloc[i]
    
    temp = re.sub('\W+', " ", text)
    temp = temp.lower()
    temp = temp.split()
    temp = [word for word in temp if not word in stopwords.words('english')]
    temp = ' '.join(temp)
    
    new_dict[ind_list[i]].append(temp)
```

```python
# Lemmatizer modules
lemmatizer = WordNetLemmatizer()

for pair in new_dict.items():
    new_dict[pair[0]] = [lemmatizer.lemmatize(answer) for answer in pair[1]]

result_df = pd.DataFrame.from_dict(new_dict, orient='index').T.stack()

result_df = result_df.reset_index(level=0, drop=True)

result_df = result_df.sort_index()
```

```python
df1 = result_df
df1 = pd.DataFrame(df1, columns = ['mention1'])
```

```python
# delete the row with emtpy values
data = df1.copy()
data.rename(columns = {'mention1': 'response'}, inplace = True)

data['id_case'] = data.index
data.reset_index(drop = True, inplace = True)

for x in range(len(data)): 
    if data['response'].iloc[x] == '': # Check if there's empty strings values
        print(x)
data
```

```python
data.drop(index = [433,945,1446,6492], inplace = True) # dropped the empty string valued rows
data.set_index('id_case', inplace=True)
df1 = data.copy()
```

```python
df1.rename(columns = {'response': 'mention1'}, inplace = True)
```

## Stack the responses

```python
# call the dataframe
df1 = pd.read_json('~/thesis/data/processed_uscensus/political_mention1.jsonl', orient='index')
df2 = pd.read_json('~/thesis/data/processed_uscensus/political_mention2.jsonl', orient='index')
df3 = pd.read_json('~/thesis/data/processed_uscensus/political_mention3.jsonl', orient='index')
```

```python
df1
```

```python
# Vertically stacking the DataFrames
merged_df= pd.concat([df1['mention1'], df2['mention2'], df3['mention3']], axis=0).reset_index(drop=True)

merged_df.info()
```

```python
merged_df = merged_df.fillna('')
```

```python
merged_df = pd.DataFrame(merged_df, columns = ['stack'])
```

```python
# delete the row with empty values
data = merged_df.copy()
data.reset_index(drop = True, inplace = True)

for x in range(len(data)): 
    if data['stack'].iloc[x] == '': # Check if there's empty strings values
        print(x)
data
```

There is no empty values after the concatenation

```python
merged_df.to_json('~/thesis/data/processed_uscensus/political_mentions_stack.jsonl', orient='records', lines = True)
```
