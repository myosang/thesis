## GloVe

without finding the closest words if they don't exist in the pre-trained model embeddings.

```python
from nltk.tokenize import word_tokenize
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
# Function to load the pre-trained word embeddings from a file
def load_embeddings(file_path):
    embeddings_dict = {}
    with open(embeddings_file_path, 'r', encoding='utf-8') as f:
        for line in f:
            values = line.split()
            word = values[0]
            vector = np.asarray(values[1:], dtype='float32')
            embeddings_dict[word] = vector
    return embeddings_dict

# Function to get the sentence embedding by averaging word embeddings
def get_sentence_embedding(sentence, embeddings_dict):
    words = word_tokenize(sentence)
    word_embeddings = [embeddings_dict[word] for word in words if word in embeddings_dict]
    if not word_embeddings:
        return np.zeros(300)  # Embedding size is 300, because the pre-trained model is based on 300 dimensions vector
    return np.mean(word_embeddings, axis=0)

# Path to the pre-trained embeddings file
embeddings_file_path = 'glove-42B-300d.txt'
embeddings_dict = load_embeddings(embeddings_file_path)
```

```python
import torch

# Check if the GPU is available
use_cuda = torch.cuda.is_available()
use_cuda
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from scipy.spatial.distance import cdist
from sklearn.decomposition import PCA
```

```python
from transformers import DistilBertTokenizer, DistilBertModel
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
model = DistilBertModel.from_pretrained("distilbert-base-uncased")
```

### Data: mention1

```python
data = pd.read_json('~/thesis/data/processed_uscensus/political_mention1.jsonl', orient='index')
```

```python
data = data["mention1"].copy()
```

```python
# Example list of texts to cluster
texts = [answer for answer in data]
```

```python
texts
```

```python
# Tokenize and encode the texts
encoded_texts = tokenizer(texts, padding=True, truncation=True, return_tensors="pt")

# Obtain RoBERTa embeddings
with torch.no_grad():
    model_output = model(**encoded_texts)

# Get the embeddings from the model output
embeddings = model_output.last_hidden_state[:, 0, :].numpy()
```

```python
embeddings.shape
```
