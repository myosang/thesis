# HDBSCAN

```python
import pandas as pd
import hdbscan as hd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import data
emb = pd.read_json('~/thesis/embeddings/response_distilbert_mentions.jsonl', orient = 'index')
```

```python
emb.head(5)
```

```python
# "transform embedding"
from ast import literal_eval
emb['embeddings'] = emb['embeddings'].apply(np.array)  # convert list to numpy array
matrix = np.vstack(emb['embeddings'].values)
matrix.shape
```

```python
X = matrix.copy()
```

```python
model = hd.HDBSCAN(min_cluster_size = 10, min_samples = 10)
```

```python
labels = model.fit_predict(X)
```

```python
model.condensed_tree_.plot(select_clusters = False)

np.unique(labels)
```

## scatter plot
make another way to visualize the hdbscan clustering result

```python
import pickle

# first change the embeddings to the array format
emb['embeddings'] = emb['embeddings'].apply(np.array) 
```

```python
with open('mentions.pkl', 'wb') as file: 
    pickle.dump(emb, file)
```

```python
with open('mentions.pkl', 'rb') as file:
    loaded_data = pickle.load(file)
    loaded_embeddings = loaded_data['embeddings']
    loaded_text = loaded_data['mentions']
```

```python
matrix = np.vstack(loaded_embeddings.values)
```

```python
import umap
reducer = umap.UMAP()
embedding_2d = reducer.fit_transform(matrix)
```

```python
# Plotting
plt.figure(figsize=(10, 8))
scatter = plt.scatter(embedding_2d[:, 0], embedding_2d[:, 1], c=labels, cmap='Spectral', alpha=0.7, edgecolors='w', s=50)
plt.colorbar(scatter, label='Cluster Label')
plt.title('HDBSCAN Clustering Results')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.show()
```

## understanding of the result HDBSCAN
- the 2d graph has the datapoints of embeddings after the umap dimensionality reduction. 
- the 2d graph shows the dimensions that are from the embeddings, which are not appropriate for interpretation. This is because of the umap algorithms, which puts the high-dimensional space into the lower-dimensional space while preserving the local relationships between the data points as well as capturing non-linear relationships between the datapoints. Therefore, it is not clear as PCA for what does each dimension mean. (ref: "UMAP: Uniform ..." from Zotero)
